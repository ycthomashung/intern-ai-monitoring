```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rt = [0.93, 0.931, 0.884, 0.83]
ut = [0.922, 0.934, 0.944, 0.84]
rs = [0.89, 0.896, 0.886, 0.788]
us = [0.89, 0.884, 0.848, 0.755]
x = ["5", "10", "25", "100"]

plt.figure(figsize=(12, 10))
plt.axhline(y=0.925, color="red", label="Original model (baseline)")
plt.plot(x, rt, color="blue", label="Random Selection + Transfer Learning")
plt.plot(x, ut, color="black", label="Uncertainty Selection + Transfer Learning")
plt.plot(x, rs, color="green", label="Random Selection + Semi-supervised Learning")
plt.plot(x, us, color="purple", label="Uncertainty Selection + Semi-supervised Learning")
plt.xlabel("Selection percentage")
plt.ylabel("Accuracy")
plt.title("Accuracy of adjusted model on training / validation data")
plt.legend()
plt.show()
```
